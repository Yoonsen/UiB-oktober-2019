## Find a book


```python
Collett = nb.book_corpus(title='amtman%døt%', author='collet%', limit=10)
Collett
```

#### Put the book into a variable

Use the value of a URN in the left column

```python
book = 2011070620007
```

```python
nb.urn_concordance(word=['hun','henne'], urns=book, size=20, before=12, after=12)
```

## Fetch a frequency list

```python
book_freq = nb.frame(nb.get_freq(book, top=10000, cutoff=4), 'freq')
book_freq.head()
```

```python
tot = nb.frame(nb.totals(50000), 'tot')
tot.head()
```

## Association

What is typical for the book?

```python
book_assoc = nb.frame_sort(nb.frame(book_freq.freq**1.5/tot.tot))
book_assoc.head(50)
```

```python
graf = nb.make_network_graph(book, list(book_assoc.head(50).index))
```

```python
gnl.show_graph(graf)
```

```python
gnl.show_communities(graf)
```

Use the community structure or build your own wordbag.

```python
book_word_bags = gnl.community_dict(graf)
book_word_bags
```

```python
d = nb.plot_book_wordbags(book, book_word_bags, window=10000, pr=500)
```

```python
d.plot(figsize=(15,5), lw=4, alpha=0.8)
```
